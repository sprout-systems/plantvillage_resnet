```python
import torch
import torchvision.transforms as transforms
from torchvision import datasets
from torch.utils.data import DataLoader, Subset
import random
from collections import defaultdict

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

dataset_path = "/root/.cache/kagglehub/datasets/emmarex/plantdisease/versions/1/PlantVillage"

full_data = datasets.ImageFolder(root=dataset_path, transform=transform)

keep_percentage = 0.1

class_indices = defaultdict(list)
for idx, (_, label) in enumerate(full_data):
    class_indices[label].append(idx)

filtered_indices = []
for label, indices in class_indices.items():
    sample_size = max(1, int(len(indices) * keep_percentage))
    filtered_indices.extend(random.sample(indices, sample_size))

train_data = Subset(full_data, filtered_indices)

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)

num_classes = len(full_data.classes)
print(f"Number of Classes: {num_classes}")
print(f"Total training images after limiting: {len(train_data)}")
print(f"Class names: {full_data.classes}")
```

```python
class_labels = {idx: name for idx, name in enumerate(full_data.classes)}
print(class_labels)
```

```python
import torch.nn as nn
import torchvision.models as models

model = models.resnet50(pretrained=True)

for param in model.parameters():
    param.requires_grad = False

model.fc = nn.Linear(model.fc.in_features, num_classes)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.001)
```

**Train**

```python
import time
import torch

num_epochs = 3

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    start_time = time.time()

    print(f"\n🚀 Starting Epoch {epoch+1}/{num_epochs}...\n")

    for batch_idx, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        if (batch_idx + 1) % 10 == 0:
            print(f"🟢 Batch {batch_idx+1}/{len(train_loader)} - Loss: {loss.item():.4f}")

    epoch_loss = running_loss / len(train_loader)
    epoch_time = time.time() - start_time

    print(f"\n✅ Epoch [{epoch+1}/{num_epochs}] Completed - Loss: {epoch_loss:.4f} - Time: {epoch_time:.2f} sec")

print("\n🎉 Training Finished!")
```

**Evaluate**

```python
correct = 0
total = 0
model.eval()

class_labels = {idx: name for idx, name in enumerate(full_data.classes)}

with torch.no_grad():
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        actual_names = [class_labels[label.item()] for label in labels]
        predicted_names = [class_labels[pred.item()] for pred in predicted]

        print(f"Actual: {actual_names}")
        print(f"Predicted: {predicted_names}")

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Training Accuracy: {100 * correct / total:.2f}%")
```

**Predict**

```python
import torch
import requests
from PIL import Image
import torchvision.transforms as transforms

image_url = "https://plantdoctor.co.nz/assets/uploads/2016/09/cherry.jpg"
image_path = "test_image.jpg"

response = requests.get(image_url)
with open(image_path, "wb") as f:
    f.write(response.content)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

image = Image.open(image_path).convert("RGB")
image = transform(image).unsqueeze(0)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
image = image.to(device)

model.to(device)
model.eval()

with torch.no_grad():
    outputs = model(image)
    probabilities = torch.softmax(outputs, dim=1)
    confidence, predicted = torch.max(probabilities, 1)

class_labels = {idx: name for idx, name in enumerate(full_data.classes)}

threshold = 0.3
print(f"Model output shape: {outputs.shape}")
predicted_class = "No Disease" if confidence.item() < threshold else class_labels[predicted.item()]

print(f"Predicted Class: {predicted_class} (Confidence: {confidence.item():.2f})")
```
